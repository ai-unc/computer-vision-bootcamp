## Blurring an Image

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('puffle.png')
cv2_imshow(img)
```

```python
box_kernel = np.array([[1, 1, 1],
                   [1, 1, 1],
                   [1, 1, 1]]) / 9

motion_blur_kernel = np.array([[1, 0, 0],
                               [0, 1, 0],
                               [0, 0, 1]]) / 3
```

```python
kernel = box_kernel

blurred_image = cv2.filter2D(img, -1, kernel)
for i in range(3000):
  blurred_image = cv2.filter2D(blurred_image, -1, kernel)

cv2_imshow(blurred_image)
```

## Sobel Edge Kernels

```python
x_kernel = np.array([[-1, 0, 1],
                     [-2, 0, 2],
                     [-1, 0, 1]])

y_kernel = np.array([[-1, -2, -1],
                     [0, 0, 0],
                     [1, 2, 1]])
```

```python
kernel = x_kernel
x_edges = cv2.filter2D(img, -1, kernel)

cv2_imshow(x_edges)
```

```python
kernel = y_kernel
y_edges = cv2.filter2D(img, -1, kernel)

cv2_imshow(y_edges)
```

```python
cv2_imshow(x_edges + y_edges)
```

## Random Kernels

```python
emboss_kernel = np.array([[-2, -1, 0],
                          [-1, 1, 1],
                          [0, 1, 2]])
```

```python
kernel = emboss_kernel

altered_image = cv2.filter2D(img, -1, kernel)

cv2_imshow(altered_image)
```
